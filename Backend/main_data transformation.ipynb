```python
df = pd.read_csv('data/normalized_ingredients_chatgpt.csv')
print(f"df.shape {df.shape}")

df.head()
```

```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer

# Example: assuming your DataFrame is named df
# First, split the "Normalized Ingredients" into lists
df['Normalized Ingredients'] = df['Normalized Ingredients'].apply(lambda x: [i.strip() for i in x.split(',')])

# Apply one-hot encoding
mlb = MultiLabelBinarizer()
one_hot = pd.DataFrame(mlb.fit_transform(df['Normalized Ingredients']), columns=mlb.classes_)

# Combine with original df if needed
df_encoded = df.drop(columns=['Normalized Ingredients']).join(one_hot)

# View result
df_encoded.head()
```

```python
df_encoded.to_csv('data/data_lebanese-dishes.csv', index=False)
```
